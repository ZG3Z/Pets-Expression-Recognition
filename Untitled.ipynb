```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
%matplotlib inline
import shutil
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.callbacks import EarlyStopping

from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D, BatchNormalization
from keras.regularizers import l1 
```

```python
main_dir = 'data/' 
img_rows, img_cols = 224, 224
```

```python
file_paths = []
labels = []

for category in os.listdir(main_dir):
    category_path = os.path.join(main_dir, category)
    if os.path.isdir(category_path):
        files = os.listdir(category_path)
        file_paths.extend([os.path.join(category_path, file) for file in files])
        labels.extend([category] * len(files))
        
data = pd.DataFrame({'file_path': file_paths, 'label': labels})
```

```python
train_data, valid_data = train_test_split(data, test_size=0.2, random_state=42)
valid_data, test_data = train_test_split(valid_data, test_size=0.5, random_state=42)
```

```python
train_data
```

```python
train_data.label.value_counts().plot(kind='bar')
```

```python
def preprocess(dataframe, target_size, column_name):
    images = []

    for index, row in dataframe.iterrows():
        path = row[column_name] 
        image = Image.open(path)
        image = image.resize(target_size)
        image_array = np.array(image)
        images.append(image_array)

    x = np.array(images)
    y = np.array(dataframe['label'])
    return x, y
```

```python
X_train, y_train = preprocess(train_data, (224, 224), 'file_path')
X_valid, y_valid = preprocess(valid_data, (224, 224), 'file_path')
X_test, y_test = preprocess(test_data, (224, 224), 'file_path')
```

```python
print(X_train.shape)
print(X_valid.shape)
print(X_test.shape)
```

```python
plt.figure(figsize=(10, 10))

for idx in range(9):
    plt.subplot(3, 3, idx + 1)
    plt.imshow(X_train[idx])
    plt.title(y_train[idx])
    plt.axis('off')

plt.tight_layout()
```

```python
input_shape = (img_rows, img_cols, 3)
input_shape
```

```python
X_train = X_train / 255
X_valid = X_valid / 255
X_test = X_test / 255
```

```python
num_classes = len(set(y_train))
label_encoder = LabelEncoder()

y_train_encoded = label_encoder.fit_transform(y_train)
y_valid_encoded = label_encoder.fit_transform(y_valid)
y_test_encoded = label_encoder.fit_transform(y_test)
```

```python
y_train_categorical = to_categorical(y_train_encoded, num_classes)
y_valid_categorical = to_categorical(y_valid_encoded, num_classes)
y_test_categorical = to_categorical(y_test_encoded, num_classes)
```

```python
y_train_categorical
```

```python
model = Sequential([
    Conv2D(filters=32, kernel_size=(3,3), input_shape=input_shape),
    BatchNormalization(),
    MaxPool2D(pool_size=(2,2)),
    
    Conv2D(filters=64, kernel_size=(3,3)),
    BatchNormalization(),
    MaxPool2D(pool_size=(2,2)),
    
    Conv2D(filters=128, kernel_size=(3,3)),
    BatchNormalization(),
    MaxPool2D(pool_size=(2,2)),
    
    Conv2D(filters=128, kernel_size=(3,3)),
    BatchNormalization(),
    MaxPool2D(pool_size=(2,2)),
    # kernel_regularizer=l1(0.01)
    Flatten(),
    Dense(128, activation='relu'),
    Dense(num_classes, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics='accuracy')
```

```python
model.summary()
```

```python
es = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
history = model.fit(
    X_train, y_train_categorical,
    batch_size=64,
    epochs=30,
    verbose=1,
    validation_data=(X_valid, y_valid_categorical),
  #  callbacks=[es]
)
```

```python
score = model.evaluate(X_test, y_test_categorical, verbose=0)
```

```python
score
```

```python
def display_training_curves(training, validation, title, subplot):
  ax = plt.subplot(subplot)
  ax.plot(training)
  ax.plot(validation)
  ax.set_title('model '+ title)
  ax.set_ylabel(title)
  ax.set_xlabel('epoch')
  ax.legend(['training', 'validation'])

plt.subplots(figsize=(10,10))
plt.tight_layout()
display_training_curves(history.history['accuracy'], history.history['val_accuracy'], 'accuracy', 211)
display_training_curves(history.history['loss'], history.history['val_loss'], 'loss', 212)
```
